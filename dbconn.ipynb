```python
import pymysql

# MariaDB 연결 정보 설정
db_host = '106.243.92.221'  # 공인 IP
db_user = 'root'
db_password = '1234'
db_name = 'nwdb'

db_port = 3306

# 데이터베이스 연결
try:
    connection = pymysql.connect(host=db_host, user=db_user, password=db_password, database=db_name, port=db_port)
    cursor = connection.cursor()

    # 여기에 SQL 쿼리 및 기타 작업을 수행합니다.

    #print("데이터베이스 연결 성공!")

    query = "SELECT * FROM news00;"
    cursor.execute(query)

    # 결과 가져오기
    result = cursor.fetchall()

    # 결과 출력
    for row in result:
        print(row)

except pymysql.MySQLError as e:
    print(f"Error: {e}")
    print("데이터베이스 연결 실패!")

finally:
    # 연결 종료
    if 'connection' in locals():
        connection.close()
```
